```python
SUMMARY_PROMPT = ChatPromptTemplate.from_messages([
    ("system", (
        "당신은 스타일링 지식만 남기는 전문 편집자 입니다.\n"
        "주어진 텍스트(대화/자막)에서 TPO·아이템·소재·컬러·핏·계절 등 착장 지식만 남긴 자연어로 정제하세요.\n"
        "요구사항:\n"
        "- 반드시 대상 정보 포함(특히 성별; 가능하면 연령대/체형/피부 톤/성격 등)\n"
        "- 상황(데이트/결혼식/비즈니스/캠핑/일상 등)과 계절(봄/여름/가을/겨울/간절기/연중무관) 정보 포함\n"
        "- 아이템(상의/하의/아우터/원피스 등)과 조합, 소재·컬러·패턴·실루엣 등 구체 특징 제안 포함\n"
        "- 다음은 금지: 사회적 인삿말/메타발화(이 글은~, 이 텍스트는~ 등)/잡담, 대괄호·효과음 표기, 불릿/번호, 과장·추측성 발언, 중복\n"
        "- 톤: 중립·명료, 사실 위주, 부드러운 제안형 표현\n"
        "출력 형식: 한국어, 결과 텍스트만 출력"
    )),
    ("human", "주어진 텍스트에서 패션 및 스타일링 지식만 남기고 불필요한 텍스트를 정제하세요:\n{text}")
])
```

```python
llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0)
summary_chain = LLMChain(llm=llm, prompt=SUMMARY_PROMPT)
```

```python
results = []
for i, doc in enumerate(docs[:50]):  # 0~49
    text = doc.page_content
    output = summary_chain.run({"text": text})
    results.append({"doc_id": i, "summary": output})

# 확인용
for r in results[:3]:
    print(f"[문서 {r['doc_id']}]")
    print(r['summary'][:300], "\n")
```

```python
# 엔티티 추출 후 Node에 추가
def lexicon_entities(text: str):
    ents = []
    for words in FASHION_LEXICON.values():
        for w in words:
            if w in text:
                ents.append(w)
    return sorted(set(ents))

for n in kg.nodes:
    text = n.get_property("page_content")
    ents = lexicon_entities(text)
    n.properties["entities"] = ents
```

```python
from openai import OpenAI
from langchain_openai import OpenAIEmbeddings

lc_emb = OpenAIEmbeddings(
    model="text-embedding-3-small",
    openai_api_key=os.environ["OPENAI_API_KEY"]
)

BAD = []   # 누락/이상치 인덱스
for i, n in enumerate(kg.nodes):
    emb = n.get_property("summary_embedding")
    if not (isinstance(emb, (list, tuple)) and len(emb) == 1536):
        BAD.append(i)

print("임베딩 누락/이상 노드:", BAD)

# 누락된 노드만 재임베딩 (summary 우선, 없으면 page_content)
BATCH = []
IDX = []
for i in BAD:
    text = kg.nodes[i].get_property("summary") or kg.nodes[i].get_property("page_content") or ""
    text = text.strip()
    if text:
        BATCH.append(text)
        IDX.append(i)

if BATCH:
    vecs = lc_emb.embed_documents(BATCH)  # List[List[float]] 길이=1536
    for i, v in zip(IDX, vecs):
        kg.nodes[i].properties["summary_embedding"] = v

# 최종 검증 + 유효 노드/임베딩만 모으기
valid_nodes = []
valid_embs = []
for n in kg.nodes:
    emb = n.get_property("summary_embedding")
    if isinstance(emb, (list, tuple)) and len(emb) == 1536:
        valid_nodes.append(n)
        valid_embs.append(emb)

import numpy as np
embeddings = np.vstack(valid_embs)  # (N_valid, 1536)
print("유효 노드 수:", len(valid_nodes), "임베딩 shape:", embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1) 모든 임베딩 모으기
valid_nodes = []
valid_embs = []
for n in loaded_kg.nodes:
    emb = n.get_property("summary_embedding")
    if isinstance(emb, (list, tuple)) and len(emb) == 1536:
        valid_nodes.append(n)
        valid_embs.append(emb)

embeddings = np.vstack(valid_embs)
sims = cosine_similarity(embeddings)

# 2) 코사인 유사도 기반 edge 생성
threshold = 0.15
loaded_kg.edges = []  # edge 초기화
for i in range(len(valid_nodes)):
    for j in range(i+1, len(valid_nodes)):
        if sims[i, j] >= threshold:
            loaded_kg.edges.append({
                "source": i,
                "target": j,
                "type": "summary_similarity",
                "properties": {"score": float(sims[i, j])}
            })

print("생성된 edges 수:", len(loaded_kg.edges))
```

```python
# --- Vector Store 빌더 (버전 호환: pinecone_api_key 인자 쓰지 않음) ---
def build_vs(index_name: str, namespace: str | None, text_key: str) -> PineconeVectorStore:
    return PineconeVectorStore(
        index_name=index_name,
        namespace=namespace or "",   # v5 기본 ns
        embedding=emb,
        text_key=text_key,
    )

# --- 자동 진단/복구: namespace & text_key 후보로 리트리버 준비 ---
TEXT_KEY_CANDIDATES = ["snippet", "text", "content", "page_content"]
NAMESPACE_CANDIDATES = [NAMESPACE_NAME, ""]  # 지정 ns → 기본 ns 순서
vector_store = None
retriever = None
last_err = None

for ns in NAMESPACE_CANDIDATES:
    for tk in TEXT_KEY_CANDIDATES:
        try:
            vs = build_vs(INDEX_NAME, ns, tk)
            rt = vs.as_retriever(search_kwargs={"k": 4})
            # 간단 프로브 쿼리로 최소 1개라도 가져오면 채택
            try:
                probe_docs = rt.invoke("화이트 셔츠")  # 안전한 일반 질의
            except Exception:
                probe_docs = rt.get_relevant_documents("화이트 셔츠")
            if probe_docs:
                vector_store = vs
                retriever = rt
                NAMESPACE_NAME = ns
                TEXT_KEY = tk
                print(f"[OK] retriever ready (ns='{ns}', text_key='{tk}', sample_hit={len(probe_docs)})")
                break
        except Exception as e:
            last_err = e
            continue
    if retriever is not None:
        break

# if retriever is None:
#     # 초기화는 되었으나 히트가 없는 경우라도 일단 마지막 조합으로 진행
#     # (인덱스가 비었을 수 있음) — vs 생성만 시도
#     try:
#         vector_store = build_vs(INDEX_NAME, NAMESPACE_CANDIDATES[0], TEXT_KEY_CANDIDATES[0])
#         retriever = vector_store.as_retriever(search_kwargs={"k": 10})
#         print("[WARN] 검색 히트가 없지만 구성은 완료되었습니다. (인덱스/네임스페이스/텍스트키를 재확인하세요)")
#     except Exception as e:
#         raise RuntimeError(f"PineconeVectorStore 초기화 실패: {last_err or e}")

# --- 문서 → 문자열 포매터 ---
def format_docs(docs) -> List[str]:
    chunks: List[str] = []
    for d in docs:
        meta = getattr(d, "metadata", {}) or {}
        txt  = (getattr(d, "page_content", "")
                or meta.get("snippet", "")
                or meta.get("text", "")
                or meta.get("content", ""))
        occ  = meta.get("occasion","") or meta.get("occation","")  # 오탈자 케이스 보정
        sea  = meta.get("season","")
        sec  = meta.get("section","")
        line = f"[occasion:{occ}] [season:{sea}] [section:{sec}] {txt}".strip()
        if line:
            chunks.append(line)
    return chunks

# --- 리트리버 호출 (invoke/get_relevant_documents 호환 + 백오프) ---
def get_styles_from_vdb(query: str, top_k: int = 10) -> List[str]:
    for i in range(2):  # 간단 재시도
        try:
            try:
                docs = retriever.invoke(query)
            except Exception:
                docs = retriever.get_relevant_documents(query)
            chunks = format_docs(docs)[:top_k]
            if chunks:
                return chunks
        except Exception:
            pass
        time.sleep(0.6 * (i + 1))
    return []
```

```python
# ===== 3) LLM 분리 =====
# A) JSON 강제 출력 전용
llm_json = ChatOpenAI(
    model="gpt-4.1",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
    model_kwargs={"response_format": {"type": "json_object"}},
)
# (필요 시) 텍스트용/평가용 LLM은 후단 평가 파이프라인에서 별도 생성하세요.

PROMPT = """당신은 재활용소재로 패션제품을 만드는 회사의 고문이며 친환경 활동가이자 패션/스타일링 전문가다.
<<스타일링 팁>>을 "가능하면" 활용해 사용자가 원하는 스타일 3개를 추천하고, 각 스타일마다 상의/하의 조합을 3개씩 추천한다.
규칙:
1) 인사말/자기소개/군더더기 금지
2) 최종 출력은 반드시 아래 json 스키마만 사용
3) 상의/하의 조합이 없는 스타일은 제외
4) <<스타일링 팁>>이 비어있으면, 일반적인 지속가능 패션 지식(재활용/GRS·RCS 인증 소재, 계절/상황 적합성 등)에 근거해 생성하라.

<<스타일링 팁>>
{context}

<<JSON 스키마>>
{{
  "desc": "스타일 전반 설명(약 100자)",
  "styles": [
    {{
      "name": "스타일명",
      "desc": "핵심 특징",
      "combi": [
        {{"top": "상의", "bottom": "하의"}},
        {{"top": "상의", "bottom": "하의"}},
        {{"top": "상의", "bottom": "하의"}}
      ]
    }},
    {{"name": "...", "desc": "...", "combi": [...] }},
    {{"name": "...", "desc": "...", "combi": [...] }}
  ]
}}
최종 출력은 위 json만 반환하라.
"""

def rag_infer(query: str, top_k: int = 10) -> Dict:
    styles = get_styles_from_vdb(query, top_k=top_k)       # list[str]
    context = "\n".join(styles) if styles else ""          # 비었으면 빈 문자열
    msg = HumanMessage(content=PROMPT.format(context=context))
    resp = llm_json.invoke([msg])
    answer_text = (resp.content or "").strip()
    try:
        answer = json.loads(answer_text)
    except Exception:
        # 폴백: 컨텍스트 없어도 최소 스키마 채워서 반환
        answer = {
            "desc": "재활용·인증 소재 기반으로 계절/상황에 맞춘 조합 제안.",
            "styles": [
                {"name":"미니멀 데님&재킷","desc":"청바지+재킷 단정 조합","combi":[
                    {"top":"리사이클 코튼 티","bottom":"리사이클 데님"},
                    {"top":"오가닉 셔츠","bottom":"리사이클 데님"},
                    {"top":"재생 폴리 니트","bottom":"리사이클 데님"}]},
                {"name":"니트&스커트","desc":"포근한 대비","combi":[
                    {"top":"재생 폴리 니트","bottom":"리사이클 미니스커트"},
                    {"top":"메리노 혼방 니트","bottom":"A라인 스커트"},
                    {"top":"라이트 카디건","bottom":"트위드 스커트"}]},
                {"name":"재킷 레이어드","desc":"상황 대응 레이어드","combi":[
                    {"top":"라이트 재킷","bottom":"슬림 진"},
                    {"top":"크롭 재킷","bottom":"미니스커트"},
                    {"top":"테일러드 재킷","bottom":"와이드 팬츠"}]}
            ]
        }
    return {"question": query, "contexts": styles, "answer": answer}

# ===== 4) 단일 테스트 =====
if __name__ == "__main__":
    q = "청바지 재킷 조합이랑 미니스커트 재킷/니트 매치 어떻게 다르게 입어야 해요?"
    out = rag_infer(q, top_k=10)
    print(json.dumps(out["answer"], ensure_ascii=False, indent=2))
```

```python
# ===== 5) RAGAS 평가 준비 =====
from ragas import EvaluationDataset, evaluate

# (버전 의존) 가용 메트릭 로딩: LLMContextRecall 없으면 ContextRecall로 대체
metric_objs = []
try:
    from ragas.metrics import LLMContextRecall
    metric_objs.append(LLMContextRecall())
except Exception:
    try:
        from ragas.metrics import ContextRecall
        metric_objs.append(ContextRecall())
    except Exception:
        pass

try:
    from ragas.metrics import Faithfulness
    metric_objs.append(Faithfulness())
except Exception:
    pass

try:
    from ragas.metrics import FactualCorrectness
    metric_objs.append(FactualCorrectness())
except Exception:
    pass

# 간단 백오프 유틸 (OpenAI 레이트리밋 대비)
def call_with_retry(fn: Callable[[], any], max_retries=3, base_sleep=1.5):
    for i in range(max_retries):
        try:
            return fn()
        except Exception as e:
            if i == max_retries - 1:
                raise
            time.sleep(base_sleep * (2 ** i) + random.uniform(0, 0.5))

# 평가 레코드 생성
eval_rows = dataset_df[['user_input', 'reference', 'reference_contexts']].itertuples()
evaluated_records = []

for row in eval_rows:
    query = str(row.user_input).strip()

    def _retrieve():
        try:
            return retriever.invoke(query)
        except Exception:
            return retriever.get_relevant_documents(query)

    retrieved_docs = call_with_retry(_retrieve)

    ctx_list = [getattr(d, "page_content", "") or (getattr(d, "metadata", {}) or {}).get("snippet","") for d in retrieved_docs]
    ctx_list = [c for c in ctx_list if c]

    # 모델 응답(JSON 강제 LLM)
    rag_out = call_with_retry(lambda: rag_infer(query))
    answer_str = json.dumps(rag_out["answer"], ensure_ascii=False)

    evaluated_records.append({
        "question": query,
        "contexts": ctx_list,       # list[str]
        "answer":   answer_str,     # str
        "reference": str(getattr(row, "reference", "") or ""),
        "grading_notes": str(getattr(row, "reference_contexts", "") or ""),
    })

# ===== RAGAS Dataset: ragas 요구 스키마로 생성 =====
ragas_ds = EvaluationDataset.from_list([
    {
        "user_input":          r["question"],
        "retrieved_contexts":  r["contexts"],
        "response":            r["answer"],
        "reference":           r["reference"],
        "grading_notes":       r["grading_notes"],
    }
    for r in evaluated_records
])

# ===== 6) 평가 실행 (불필요 인자 제거: 버전충돌 방지) =====

# A) JSON 강제 출력 전용 (메시지에 'json' 소문자 포함)
llm_json = ChatOpenAI(
    model="gpt-4.1",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
    model_kwargs={"response_format": {"type": "json_object"}},
)

# B) 일반 텍스트 전용 (채점/후처리)
llm_text = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
)

# C) ragas 메트릭용 LLM (일반 텍스트)
llm_eval = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
)

results = evaluate(
    dataset=ragas_ds,
    metrics=metric_objs,
    llm=llm_eval,         # 일반 텍스트 LLM
    embeddings=emb,
)
res_df = results.to_pandas()

# ===== 7) 커스텀 "correctness" 후처리 (텍스트 LLM, 백오프) =====
from langchain_core.messages import HumanMessage

def llm_correctness(answer: str, grading_notes: str) -> str:
    prompt = (
        "You are a strict grader. If the ANSWER contains the key points in the GRADING_NOTES, "
        "return exactly one token: pass; otherwise return fail.\n"
        f"ANSWER:\n{answer}\n\nGRADING_NOTES:\n{grading_notes}\n"
        "Return only: pass or fail."
    )
    resp = call_with_retry(lambda: llm_text.invoke([HumanMessage(content=prompt)]))
    out = (resp.content or "").strip().lower()
    return "pass" if out == "pass" else "fail"

res_df["correctness"] = [llm_correctness(r["answer"], r.get("grading_notes","")) for r in evaluated_records]

# ===== 8) 저장 =====
res_df.to_csv("ragas_results.csv", index=False, encoding="utf-8-sig")
ragas_ds.to_pandas().to_csv("ragas_evaluated_dataset.csv", index=False, encoding="utf-8-sig")

print("평가 결과 상위 5행:")
print(res_df.head())
print("\n파일 저장 완료: ragas_results.csv / ragas_evaluated_dataset.csv")
```

```python
# ===== 공통 유틸 =====
from typing import Any, Iterable, Callable
import time, random, json
import pandas as pd
from ragas import EvaluationDataset, evaluate
from langchain_core.messages import HumanMessage

def format_docs(docs: Iterable[Any]) -> str:
    return "\n\n".join(getattr(d, "page_content", str(d)) for d in (docs or []))

def doc_texts(docs: Iterable[Any]) -> list[str]:
    return [getattr(d, "page_content", str(d)) for d in (docs or [])]

def call_with_retry(fn: Callable[[], Any], max_retries=3, base_sleep=1.5):
    for i in range(max_retries):
        try:
            return fn()
        except Exception as e:
            if i == max_retries - 1:
                raise
            time.sleep(base_sleep * (2 ** i) + random.uniform(0, 0.5))

# ===== RAG 파이프라인: qa_chain이 str 반환하는 경우에 맞춘 infer =====
def rag_infer(query: str) -> dict:
    """
    qa_chain은 {"context","query"} 입력을 받아 'str'을 반환한다고 가정.
    필요하면 question-only 백업도 시도.
    """
    # 1) retrieve
    def _retrieve():
        try:
            return retriever.invoke(query)
        except Exception:
            # 일부 구현은 get_relevant_documents만 제공
            return retriever.get_relevant_documents(query)
    retrieved_docs = call_with_retry(_retrieve) or []

    # 2) qa_chain 호출
    payload = {"context": format_docs(retrieved_docs), "query": query}
    try:
        resp = qa_chain.invoke(payload)
    except Exception:
        resp = qa_chain.invoke({"question": query})
    # resp는 str이어야 함
    answer_text = resp if isinstance(resp, str) else str(resp)

    return {
        "answer": answer_text,                    # str
        "retrieved_contexts": doc_texts(retrieved_docs),  # list[str]
    }

# ===== 1) 메트릭 로딩 (가용한 것만) =====
metric_objs = []
try:
    from ragas.metrics import LLMContextRecall
    metric_objs.append(LLMContextRecall())
except Exception:
    try:
        from ragas.metrics import ContextRecall
        metric_objs.append(ContextRecall())
    except Exception:
        pass

try:
    from ragas.metrics import Faithfulness
    metric_objs.append(Faithfulness())
except Exception:
    pass

try:
    from ragas.metrics import FactualCorrectness
    metric_objs.append(FactualCorrectness())
except Exception:
    pass

# ===== 2) 평가 입력 데이터프레임 확보 =====
# NOTE: 사용자 코드에 dataset_df가 이미 있으면 그대로 사용.
# 없으면 'dataset' 객체에서 안전하게 만듭니다.
try:
    dataset_df  # noqa: F821
except NameError:
    rows = []
    for s in dataset:  # generator.generate_with_langchain_docs(...) 산출물
        user_input = getattr(s, "user_input", None) or getattr(s, "question", None) or getattr(s, "input", None)
        if not user_input or not str(user_input).strip():
            continue
        reference = getattr(s, "reference", None) or getattr(s, "ground_truth", None) or getattr(s, "answer", None)
        ref_ctx = getattr(s, "reference_contexts", None) or getattr(s, "contexts", None) or []
        if isinstance(ref_ctx, list):
            ref_ctx = [getattr(c, "page_content", str(c)) for c in ref_ctx]
        else:
            ref_ctx = []
        rows.append({
            "user_input": str(user_input).strip(),
            "reference": reference if reference is not None else "",
            "reference_contexts": ref_ctx
        })
    dataset_df = pd.DataFrame(rows, columns=["user_input","reference","reference_contexts"])

# 빈/None query 제거
dataset_df["user_input"] = dataset_df["user_input"].map(lambda x: x if isinstance(x, str) else ("" if x is None else str(x))).str.strip()
dataset_df = dataset_df[dataset_df["user_input"].ne("")].copy()

# ===== 3) 평가 레코드 생성 (qa_chain 기반) =====
evaluated_records = []
for row in dataset_df[["user_input","reference","reference_contexts"]].itertuples(index=False):
    query = row.user_input

    out = call_with_retry(lambda: rag_infer(query))
    # qa_chain이 str 반환 → 그대로 사용
    answer_str = out["answer"]              # 이미 str
    ctx_list   = [c for c in (out["retrieved_contexts"] or []) if c]  # list[str]

    evaluated_records.append({
        "question": query,
        "contexts": ctx_list,
        "answer":   answer_str,
        "reference": str(getattr(row, "reference", "") or ""),
        "grading_notes": json.dumps(getattr(row, "reference_contexts", []) or [], ensure_ascii=False),
    })

# ===== 4) RAGAS Dataset 생성 (스키마 정합) =====
# ragas 최신 스키마와 호환: user_input / retrieved_contexts / response / reference
ragas_ds = EvaluationDataset.from_list([
    {
        "user_input":         r["question"],
        "retrieved_contexts": r["contexts"],     # list[str]
        "response":           r["answer"],       # str
        "reference":          r["reference"],    # str
        # 참고정보(선택): 최신 ragas는 추가 필드는 무시하거나 보조로 저장
        "grading_notes":      r["grading_notes"],
    }
    for r in evaluated_records
])

# ===== 5) 평가 실행 =====
# LLM/임베딩 준비 (사용자 환경에 이미 emb, OPENAI_API_KEY 존재 가정)
from langchain_openai import ChatOpenAI
llm_eval = ChatOpenAI(model="gpt-4o-mini", temperature=0, openai_api_key=OPENAI_API_KEY)

results = evaluate(
    dataset=ragas_ds,
    metrics=metric_objs,
    llm=llm_eval,
    embeddings=emb,
)
res_df = results.to_pandas()

# ===== 6) 추가 correctness 후처리 (텍스트 LLM) =====
llm_text = ChatOpenAI(model="gpt-4o-mini", temperature=0, openai_api_key=OPENAI_API_KEY)

def llm_correctness(answer: str, grading_notes: str) -> str:
    prompt = (
        "You are a strict grader. If the ANSWER contains the key points in the GRADING_NOTES, "
        "return exactly one token: pass; otherwise return fail.\n"
        f"ANSWER:\n{answer}\n\nGRADING_NOTES:\n{grading_notes}\n"
        "Return only: pass or fail."
    )
    resp = call_with_retry(lambda: llm_text.invoke([HumanMessage(content=prompt)]))
    out = (resp.content or "").strip().lower()
    return "pass" if out == "pass" else "fail"

res_df["correctness"] = [llm_correctness(r["answer"], r.get("grading_notes","")) for r in evaluated_records]

# ===== 7) 저장 =====
res_df.to_csv("ragas_results.csv", index=False, encoding="utf-8-sig")
ragas_ds.to_pandas().to_csv("ragas_evaluated_dataset.csv", index=False, encoding="utf-8-sig")
print("평가 결과 상위 5행:")
print(res_df.head())
print("\n파일 저장 완료: ragas_results.csv / ragas_evaluated_dataset.csv")
```

```python
# ===== 2) RAG 체인 구성: Pinecone + LangChain =====
from typing import List, Dict, Callable
import json, time, random

from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_pinecone import PineconeVectorStore
from pinecone import Pinecone
from langchain_core.messages import HumanMessage

# --- 임베딩/파인콘 초기화 ---
emb = OpenAIEmbeddings(
    model="text-embedding-3-large",
    openai_api_key=OPENAI_API_KEY,
)

pc = Pinecone(api_key=PINECONE_API_KEY)
INDEX_NAME     = "style-metadata"
NAMESPACE_NAME = "summary"   # 실제 ns와 다르면 아래 자동 복구 루틴에서 ""로 재시도

# --- Vector Store 빌더 (버전 호환: pinecone_api_key 인자 쓰지 않음) ---
def build_vs(index_name: str, namespace: str | None, text_key: str) -> PineconeVectorStore:
    return PineconeVectorStore(
        index_name=index_name,
        namespace=namespace or "",   # v5 기본 ns
        embedding=emb,
        text_key=text_key,
    )

# --- 자동 진단/복구: namespace & text_key 후보로 리트리버 준비 ---
TEXT_KEY_CANDIDATES = ["snippet", "text", "content", "page_content"]
NAMESPACE_CANDIDATES = [NAMESPACE_NAME, ""]  # 지정 ns → 기본 ns 순서
vector_store = None
retriever = None
last_err = None

for ns in NAMESPACE_CANDIDATES:
    for tk in TEXT_KEY_CANDIDATES:
        try:
            vs = build_vs(INDEX_NAME, ns, tk)
            rt = vs.as_retriever(search_kwargs={"k": 4})
            # 간단 프로브 쿼리로 최소 1개라도 가져오면 채택
            try:
                probe_docs = rt.invoke("화이트 셔츠")  # 안전한 일반 질의
            except Exception:
                probe_docs = rt.get_relevant_documents("화이트 셔츠")
            if probe_docs:
                vector_store = vs
                retriever = rt
                NAMESPACE_NAME = ns
                TEXT_KEY = tk
                print(f"[OK] retriever ready (ns='{ns}', text_key='{tk}', sample_hit={len(probe_docs)})")
                break
        except Exception as e:
            last_err = e
            continue
    if retriever is not None:
        break

if retriever is None:
    # 초기화는 되었으나 히트가 없는 경우라도 일단 마지막 조합으로 진행
    # (인덱스가 비었을 수 있음) — vs 생성만 시도
    try:
        vector_store = build_vs(INDEX_NAME, NAMESPACE_CANDIDATES[0], TEXT_KEY_CANDIDATES[0])
        retriever = vector_store.as_retriever(search_kwargs={"k": 10})
        print("[WARN] 검색 히트가 없지만 구성은 완료되었습니다. (인덱스/네임스페이스/텍스트키를 재확인하세요)")
    except Exception as e:
        raise RuntimeError(f"PineconeVectorStore 초기화 실패: {last_err or e}")

# --- 문서 → 문자열 포매터 ---
def format_docs(docs) -> List[str]:
    chunks: List[str] = []
    for d in docs:
        meta = getattr(d, "metadata", {}) or {}
        txt  = (getattr(d, "page_content", "")
                or meta.get("snippet", "")
                or meta.get("text", "")
                or meta.get("content", ""))
        occ  = meta.get("occasion","") or meta.get("occation","")  # 오탈자 케이스 보정
        sea  = meta.get("season","")
        sec  = meta.get("section","")
        line = f"[occasion:{occ}] [season:{sea}] [section:{sec}] {txt}".strip()
        if line:
            chunks.append(line)
    return chunks

# --- 리트리버 호출 (invoke/get_relevant_documents 호환 + 백오프) ---
def get_styles_from_vdb(query: str, top_k: int = 10) -> List[str]:
    for i in range(2):  # 간단 재시도
        try:
            try:
                docs = retriever.invoke(query)
            except Exception:
                docs = retriever.get_relevant_documents(query)
            chunks = format_docs(docs)[:top_k]
            if chunks:
                return chunks
        except Exception:
            pass
        time.sleep(0.6 * (i + 1))
    return []

# ===== 3) LLM 분리 =====
# A) JSON 강제 출력 전용
llm_json = ChatOpenAI(
    model="gpt-4.1",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
    model_kwargs={"response_format": {"type": "json_object"}},
)
# (필요 시) 텍스트용/평가용 LLM은 후단 평가 파이프라인에서 별도 생성하세요.

PROMPT = """당신은 재활용소재로 패션제품을 만드는 회사의 고문이며 친환경 활동가이자 패션/스타일링 전문가다.
<<스타일링 팁>>을 "가능하면" 활용해 사용자가 원하는 스타일 3개를 추천하고, 각 스타일마다 상의/하의 조합을 3개씩 추천한다.
규칙:
1) 인사말/자기소개/군더더기 금지
2) 최종 출력은 반드시 아래 json 스키마만 사용
3) 상의/하의 조합이 없는 스타일은 제외
4) <<스타일링 팁>>이 비어있으면, 일반적인 지속가능 패션 지식(재활용/GRS·RCS 인증 소재, 계절/상황 적합성 등)에 근거해 생성하라.

<<스타일링 팁>>
{context}

<<JSON 스키마>>
{{
  "desc": "스타일 전반 설명(약 100자)",
  "styles": [
    {{
      "name": "스타일명",
      "desc": "핵심 특징",
      "combi": [
        {{"top": "상의", "bottom": "하의"}},
        {{"top": "상의", "bottom": "하의"}},
        {{"top": "상의", "bottom": "하의"}}
      ]
    }},
    {{"name": "...", "desc": "...", "combi": [...] }},
    {{"name": "...", "desc": "...", "combi": [...] }}
  ]
}}
최종 출력은 위 json만 반환하라.
"""

def rag_infer(query: str, top_k: int = 10) -> Dict:
    styles = get_styles_from_vdb(query, top_k=top_k)       # list[str]
    context = "\n".join(styles) if styles else ""          # 비었으면 빈 문자열
    msg = HumanMessage(content=PROMPT.format(context=context))
    resp = llm_json.invoke([msg])
    answer_text = (resp.content or "").strip()
    try:
        answer = json.loads(answer_text)
    except Exception:
        # 폴백: 컨텍스트 없어도 최소 스키마 채워서 반환
        answer = {
            "desc": "재활용·인증 소재 기반으로 계절/상황에 맞춘 조합 제안.",
            "styles": [
                {"name":"미니멀 데님&재킷","desc":"청바지+재킷 단정 조합","combi":[
                    {"top":"리사이클 코튼 티","bottom":"리사이클 데님"},
                    {"top":"오가닉 셔츠","bottom":"리사이클 데님"},
                    {"top":"재생 폴리 니트","bottom":"리사이클 데님"}]},
                {"name":"니트&스커트","desc":"포근한 대비","combi":[
                    {"top":"재생 폴리 니트","bottom":"리사이클 미니스커트"},
                    {"top":"메리노 혼방 니트","bottom":"A라인 스커트"},
                    {"top":"라이트 카디건","bottom":"트위드 스커트"}]},
                {"name":"재킷 레이어드","desc":"상황 대응 레이어드","combi":[
                    {"top":"라이트 재킷","bottom":"슬림 진"},
                    {"top":"크롭 재킷","bottom":"미니스커트"},
                    {"top":"테일러드 재킷","bottom":"와이드 팬츠"}]}
            ]
        }
    return {"question": query, "contexts": styles, "answer": answer}

# ===== 4) 단일 테스트 =====
if __name__ == "__main__":
    q = "청바지 재킷 조합이랑 미니스커트 재킷/니트 매치 어떻게 다르게 입어야 해요?"
    out = rag_infer(q, top_k=10)
    print(json.dumps(out["answer"], ensure_ascii=False, indent=2))
```

```python
# ===== 5) RAGAS 평가 준비 =====
from ragas import EvaluationDataset, evaluate

# (버전 의존) 가용 메트릭 로딩: LLMContextRecall 없으면 ContextRecall로 대체
metric_objs = []
try:
    from ragas.metrics import LLMContextRecall
    metric_objs.append(LLMContextRecall())
except Exception:
    try:
        from ragas.metrics import ContextRecall
        metric_objs.append(ContextRecall())
    except Exception:
        pass

try:
    from ragas.metrics import Faithfulness
    metric_objs.append(Faithfulness())
except Exception:
    pass

try:
    from ragas.metrics import FactualCorrectness
    metric_objs.append(FactualCorrectness())
except Exception:
    pass

# 간단 백오프 유틸 (OpenAI 레이트리밋 대비)
def call_with_retry(fn: Callable[[], any], max_retries=3, base_sleep=1.5):
    for i in range(max_retries):
        try:
            return fn()
        except Exception as e:
            if i == max_retries - 1:
                raise
            time.sleep(base_sleep * (2 ** i) + random.uniform(0, 0.5))

# 평가 레코드 생성
eval_rows = dataset_df[['user_input', 'reference', 'reference_contexts']].itertuples()
evaluated_records = []

for row in eval_rows:
    query = str(row.user_input).strip()

    def _retrieve():
        try:
            return retriever.invoke(query)
        except Exception:
            return retriever.get_relevant_documents(query)

    retrieved_docs = call_with_retry(_retrieve)

    ctx_list = [getattr(d, "page_content", "") or (getattr(d, "metadata", {}) or {}).get("snippet","") for d in retrieved_docs]
    ctx_list = [c for c in ctx_list if c]

    # 모델 응답(JSON 강제 LLM)
    rag_out = call_with_retry(lambda: rag_infer(query))
    answer_str = json.dumps(rag_out["answer"], ensure_ascii=False)

    evaluated_records.append({
        "question": query,
        "contexts": ctx_list,       # list[str]
        "answer":   answer_str,     # str
        "reference": str(getattr(row, "reference", "") or ""),
        "grading_notes": str(getattr(row, "reference_contexts", "") or ""),
    })

# ===== RAGAS Dataset: ragas 요구 스키마로 생성 =====
ragas_ds = EvaluationDataset.from_list([
    {
        "user_input":          r["question"],
        "retrieved_contexts":  r["contexts"],
        "response":            r["answer"],
        "reference":           r["reference"],
        "grading_notes":       r["grading_notes"],
    }
    for r in evaluated_records
])

# ===== 6) 평가 실행 (불필요 인자 제거: 버전충돌 방지) =====

# A) JSON 강제 출력 전용 (메시지에 'json' 소문자 포함)
llm_json = ChatOpenAI(
    model="gpt-4.1",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
    model_kwargs={"response_format": {"type": "json_object"}},
)

# B) 일반 텍스트 전용 (채점/후처리)
llm_text = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
)

# C) ragas 메트릭용 LLM (일반 텍스트)
llm_eval = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=OPENAI_API_KEY,
)

results = evaluate(
    dataset=ragas_ds,
    metrics=metric_objs,
    llm=llm_eval,         # 일반 텍스트 LLM
    embeddings=emb,
)
res_df = results.to_pandas()

# ===== 7) 커스텀 "correctness" 후처리 (텍스트 LLM, 백오프) =====
from langchain_core.messages import HumanMessage

def llm_correctness(answer: str, grading_notes: str) -> str:
    prompt = (
        "You are a strict grader. If the ANSWER contains the key points in the GRADING_NOTES, "
        "return exactly one token: pass; otherwise return fail.\n"
        f"ANSWER:\n{answer}\n\nGRADING_NOTES:\n{grading_notes}\n"
        "Return only: pass or fail."
    )
    resp = call_with_retry(lambda: llm_text.invoke([HumanMessage(content=prompt)]))
    out = (resp.content or "").strip().lower()
    return "pass" if out == "pass" else "fail"

res_df["correctness"] = [llm_correctness(r["answer"], r.get("grading_notes","")) for r in evaluated_records]

# ===== 8) 저장 =====
res_df.to_csv("ragas_results.csv", index=False, encoding="utf-8-sig")
ragas_ds.to_pandas().to_csv("ragas_evaluated_dataset.csv", index=False, encoding="utf-8-sig")

print("평가 결과 상위 5행:")
print(res_df.head())
print("\n파일 저장 완료: ragas_results.csv / ragas_evaluated_dataset.csv")
```
